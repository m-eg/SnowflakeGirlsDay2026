## Der Tatort 🐻‍❄️

Snowy kommt heute Morgen ins Büro und ...

### seine Blaubeerpfannkuchen sind verschwunden!

Es gibt **6 Verdächtige**. Einer von ihnen war es.

## Schritt 1: Wer sind die Verdächtigen?

Lasst uns zuerst anschauen, wen Snowy verdächtigt.

Drückt den **▶ Play-Button** neben der grauen Zelle!

````python
%%sql -r alle_verdaechtigen
SELECT * FROM GIRLSDAY.PUBLIC.VERDAECHTIGE;
````

**6 Verdächtige!** Schaut euch die Tabelle an, was seht ihr?

Ihr habt gerade eure **erste Datenbank-Abfrage** ausgeführt.
Die Sprache, die ihr dafür benutzt habt, heißt **SQL**.

Schauen wir uns an, was die Abfrage bedeutet:

```
SELECT *          ←  Was wollen wir sehen? (* = alles)
FROM VERDAECHTIGE ←  Aus welcher Tabelle?
```

Und schaut mal die Spalten in der Tabelle an, da gibt es verschiedene **Arten von Daten**:

| Spalte | Datentyp | Bedeutung |
|--------|----------|-----------|
| Name, Tier, Lieblingssnack | **String** | Text und Wörter |
| Groesse_CM | **Integer** | Ganze Zahlen |
| Hat_Schluessel | **Boolean** | Wahr (`TRUE`) oder Falsch (`FALSE`) |

## Schritt 2: Wer hat einen Schlüssel?

Der Dieb muss einen **Schlüssel zur Küche** gehabt haben.

Aber wir wollen jetzt nicht *alle* sehen, sondern nur die **mit Schlüssel**.
Dafür gibt es ein neues Wort: `WHERE`. Das ist eine **Bedingung**.

Drückt ▶

````python
%%sql -r hinweis_1
SELECT Name, Tier
FROM GIRLSDAY.PUBLIC.VERDAECHTIGE
WHERE Hat_Schluessel = TRUE;
````

🔎 **4 Verdächtige übrig:** Frosty, Luna, Bjørn, Finn

**Ausgeschieden:** Ivy und Mika — kein Schlüssel!

Was war neu?

```
WHERE Hat_Schluessel = TRUE   ←  Zeige nur Zeilen, wo das WAHR ist
```

## Schritt 3: Wer war in der Küche?

Der Pfannkuchen 🥞 verschwand **zwischen 12:00 und 13:00 Uhr**.

Snowy hat Kamera-Aufnahmen! Diesmal brauchen wir **zwei Bedingungen**:
der Raum muss Küche sein, **und** die Uhrzeit muss zwischen 12 und 13 Uhr liegen.

Drückt ▶

````python
%%sql -r hinweis_2
SELECT Name, Raum, Uhrzeit
FROM GIRLSDAY.PUBLIC.KAMERA_LOG
WHERE Raum = 'Küche'
AND Uhrzeit BETWEEN '12:00' AND '13:00';
````

❄️ **3 Verdächtige waren zur Tatzeit in der Küche:** Luna, Bjørn, Finn

**Ausgeschieden:** Frosty

Was war neu?

```
AND              ←  Mehrere Bedingungen verknüpfen
BETWEEN x AND y  ←  Alles zwischen zwei Werten
```

Wo war Frosty dann? Lasst uns das mit SQL überprüfen:

````python
%%sql -r dataframe_1
SELECT Name, Raum, Uhrzeit
FROM GIRLSDAY.PUBLIC.KAMERA_LOG
WHERE Name = 'Frosty'
AND Uhrzeit BETWEEN '12:00' AND '13:00';
````

## Schritt 4: Wo sind die Spuren?

Am Tatort wurden **Blaubeerspuren** 🫐 gefunden!

Aber in der Datenbank steht nicht einfach "Blaubeere", es steht z.B. "Blaubeerfleck auf der Türklinke" oder vielleicht sogar ein Adjektiv "blaubeerig".
Wir brauchen eine **Textsuche**, die nach einem Teil des Wortes sucht, das funktioniert mit '%' und 'LIKE'.


````python
%%sql -r hinweis_3a
SELECT Beschreibung, Fundort
FROM GIRLSDAY.PUBLIC.BEWEISE
WHERE Beschreibung LIKE '%laubeer%';
````

Blaubeerspuren 🫐 im **Labor** gefunden!

Was war neu?

```
LIKE '%laubeer%'  ←  Suche nach Text, der "laubeer" enthält
                      % ist ein Platzhalter für "egal was davor/danach steht"
```

Jetzt die entscheidende Frage: **Wer war im Labor?**

````python
%%sql -r hinweis_3b
SELECT Name, Raum, Uhrzeit
FROM GIRLSDAY.PUBLIC.KAMERA_LOG
WHERE Raum = 'Labor';
````

Nur EINE Person war im Labor: **Luna** die Schneeeule!
